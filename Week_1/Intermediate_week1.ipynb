  **1. User-defined Functions**

 **even_odd() returns result whether number is even or odd**

```
def even_odd(a):
  if a%2 == 0:
    return "Even"
  return "Odd"
x = 5
y = 10
print("X is ",even_odd(x),"and Y is",even_odd(y))
```

**Map():**

**Map() maps the array of inputs to a given function to produce an output for every element passed to it.**

```
def double(a):
  return a*2
num_list = [1,2,3,4]
print(list(map(double,num_list)))
```

**split():**

**It splits the input based on a particular delimiter, ',' in this case**

```
test_String = 'Python,Programming'
print(test_String.split(','))
```

**lambda():**

**lambda() is an anonymous function which returns the value of the function by default.** 

```
num_list = [2,4,6,8]
result = list(map(lambda x: x*x, num_list))
print(result)
```

**lambda() + filter():**

**filter will filter out even and odd numbers with help of lambnda function**

```
num_list = [43,22,65,12,10,77]
print("Even List:",list(filter(lambda x: x%2!=1,num_list)))
print("Odd List:",list(filter(lambda x: x%2!=0,num_list)))
```

**2. Performing basic operations on Numpy Arrays:**

**This section includes array creation, various array operations, indexing and slicing using Numpy functionalities.**

**The appropriate comments are included for each operation.**

```
import numpy as np
# a. Creating arrays
x1 = np.array((1,2,3))
print("Using tuple:",x1)
x2 = np.array(range(10))
print("Using range:",x2)
x3 = np.random.random(3)
print("A Random array:",x3)

# b. Array Operations
x4 = [[1,2,3],[4,5,6]]
print("Element at an index:",x4[1][1]) #Array Indexing
x6 = [0,1,2,3,4,5,6]
print("Sliced Array",x6[0:3]) #Array Slicing
x5 = [10,20,30,40,50] 
print("Reversed array:",x5[::-1]) 
x4[1][1] = 8 #Array Manipulation
print("Manipulated Value:",x4)

# Saving and loading numpy array
np.save('abc',x5)   # Saving the array
file = np.load('abc.npy')
print("Array loading from local disk:",file)



```

```
num_list = [10,20,30,40,50]  # Basic array creation
print(pd.Series(num_list))   # pd.Series() creates a series
```

**  2. Dataframe:**

  **Pandas Dataframe is a two-dimensional size-mutable and heterogeneous structure with labelled axes with rows and columns.**
  
  **This section shows dataframe creation along with operations on it.**

```
import pandas as pd
data = [[1,0,0],[0,1,0],[0,0,1]]  # Dataframe using list
dataframe = pd.DataFrame(data)
print(dataframe)
```

```
# Dataframe operations 
data = {'Degree':['BCA','MCA','BE','BTech'],'Seats':[60,50,60,50]}
dataframe = pd.DataFrame(data)  # Dataframe using dictionary of lists
print(dataframe)

print(dataframe['Degree']) #Accessing a column
print(dataframe[0:2]) #Accessing Elements

dataframe['Seats Filled'] = pd.Series([52,59,60,55], index=[0,1,2,3]) #Adding a column
print(dataframe)

del dataframe['Seats Filled'] #Deleting a column
print(dataframe)
```

```
# Loading, saving and displaying image
from matplotlib import pyplot as plt
import cv2

img = cv2.imread('test.jpg')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Returns an image

plt.imshow(rgb)  # Displays the image
plt.title('flowers')
plt.show()
```

  **b. Editing and saving images:**

```
import cv2
from matplotlib import pyplot as plt

ksize = (10,10)
img2 = cv2.blur(img,ksize) # Image is blurred a little bit.
plt.imshow(img2)
plt.show()
```

```
img[img[...,0]>150] = [255,255,255]

# Save result
cv2.imwrite('result.jpg', img)
plt.imshow(img)
plt.show()
```

```
import numpy as np
new_shaped_image = np.rollaxis(img, axis=1, start=0)
plt.imshow(new_shaped_image)
plt.show()
```

```
x,y,h,w = 100,100,200,300
crop_img = img[y:y+h, x:x+w]
plt.imshow(crop_img)
plt.show()
```
